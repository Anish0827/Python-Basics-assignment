Q5. Reverse a number using while loop.

Approach:
- Extract last digit
- Build reversed number

```python
#write answer here

var= int(input())

new = var

res=0

while var>0:
  temp = var % 10
  res = res*10 + temp
  var = var // 10

print(res)


if new==res:
  print("Palindrome")
else:
  print("Not Palindrome")


```

Q6. Count digits in a number.

Approach:
- Divide number by 10 repeatedly
- Count iterations

```python
#write answer here
n=int(input())
count=0
while n>0:
  n=n//10
  count=count+1
print(count)
```

Q7. Check if a number is palindrome.

```python
#write answer here

n=121
if str(n)==str(n)[-1::-1]:
  print("Palindrome")
else:
  print("Not Palindrome")
```
